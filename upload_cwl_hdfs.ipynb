```python
from hdfs import InsecureClient
```

```python
client = InsecureClient('http://namenode:50070', user = 'vagrant')
```

```python
import os 

for foldername in os.listdir('./'):
    if foldername.startswith('structured'):
        localpath = os.path.join(foldername)
        hdfspath = os.path.join('/Users/vagrant/',foldername)
        client.upload(hdfspath,localpath)
        
```

```python
# testing if all the folders got uploaded
client.list('/Users/vagrant/')
```
